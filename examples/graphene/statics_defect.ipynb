```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
df = pd.read_csv("vacancy.csv")
energy_orders = np.sort(df.energy_order.unique())
```

```python
ratios = []
energys = []
for energy_order in energy_orders:
    tdf = df[df.energy_order == energy_order]
    vs = np.sum(tdf.nVacancies)
    ratio = vs/100000
    ratios.append(ratio)
    energys.append(10*1.3**energy_order)
```

```python
plt.plot(energys[1:], ratios[1:],'.-')
plt.xscale('log')
```

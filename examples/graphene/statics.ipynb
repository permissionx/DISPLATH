```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv("p.debug.log")
```

```python
plt.plot(df.nrun, df.ntargets)
```

```python
df
```

```python
dpx = df.px0 - df.px1
dpy = df.py0 - df.py1
dpz = df.pz0 - df.pz1
```

```python
fig, axs = plt.subplots(3,1,figsize=(10,8))
for ax, data in zip(axs, [dpx,dpy, dpz]):
    ax.hist(data, bins = 50, density = True)
    ax.set_yscale("log")
```

```python
fig, axs = plt.subplots(1,3,figsize=(13,4))
for ax, xdata, ydata in zip(axs, [df.px0, df.py0, df.pz0],[dpx,dpy, dpz]):
    ax.scatter(np.abs(xdata), np.abs(ydata))
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("p.debug.log")
```

```python
plt.plot(df.nrun, df.ntargets)
```

```python
dpx = df.px1 - df.px0
dpy = df.py1 - df.py0
dpz = df.pz1 - df.pz0

fig, axs = plt.subplots(3, 1, figsize=(10, 8))
for dp, ax in zip([dpx, dpy, dpz], axs):
    ax.hist(dp, bins=100, density=True)
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_csv("p_debug.csv")
```

```python
dpx = df.p0x - df.p1x
dpy = df.p0y - df.p1y
dpz = df.p0z - df.p1z
```

```python
fig, axs = plt.subplots(3,1,figsize=(10,8))
for ax, data in zip(axs, [dpx,dpy, dpz]):
    ax.hist(data, bins = 50, density = True)
    ax.set_yscale("log")
```

```python
fig, axs = plt.subplots(1,3,figsize=(13,4))
for ax, xdata, ydata in zip(axs, [df.p0x, df.p0y, df.p0z],[dpx, dpy, dpz]):
    ax.scatter(np.abs(xdata), np.abs(ydata))
```
